# Basic Imports

```python
import numpy as np
import scipy.io as sio
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
%matplotlib inline
```

# Support Vector Machines

# Example 1 - Linear SVM

```python
#load data
data1=sio.loadmat('./ex6/ex6data1')
x=data1['X']
y=data1['y']
```

```python
def plotdata(x,y):
    data_comb=np.c_[x,y]
    pos=data_comb[data_comb[:,2]==1]
    neg=data_comb[data_comb[:,2]==0]
    plt.scatter(pos[:,0],pos[:,1],marker='+',color='black',label='Positive')
    plt.scatter(neg[:,0],neg[:,1],marker='o',color='yellow',label='Negative',edgecolors='black')
    plt.xlim(0,4.5)
    plt.ylim(1.5,5)
    plt.show()
plotdata(x,y)
```

```python
(xmin, xmax, ymin, ymax)=(0,4.5,1.5,5)
```

```python
def plotBoundary(my_svm):
    """
    Function to plot the decision boundary for a trained SVM
    It works by making a grid of x1 ("xvals") and x2 ("yvals") points,
    And for each, computing whether the SVM classifies that point as
    True or False. Then, a contour is drawn with a built-in pyplot function.
    """
    xvals = np.linspace(xmin,xmax,200)
    yvals = np.linspace(ymin,ymax,200)
    zvals = np.zeros((len(xvals),len(yvals)))
    for i in range(len(xvals)):
        for j in range(len(yvals)):
            x1=np.c_[xvals[i],yvals[j]]
            zvals[i][j] = float(my_svm.predict(x1))
    zvals=zvals.transpose()
    data_comb=np.c_[x,y]
    pos=data_comb[data_comb[:,2]==1]
    neg=data_comb[data_comb[:,2]==0]
    plt.figure(figsize=(10,6))
    plt.scatter(pos[:,0],pos[:,1],marker='+',color='black',label='Positive')
    plt.scatter(neg[:,0],neg[:,1],marker='o',color='yellow',label='Negative',edgecolors='black')
    plt.xlim(0,4.5)
    plt.ylim(1.5,5)
    u, v = np.meshgrid( xvals, yvals )
    myc=plt.contour( u, v, zvals)
```

```python
svm_lin=SVC(C=1,kernel='linear')
svm_lin.fit(x,y.flatten())
```

```python
plotBoundary(svm_lin)
```
